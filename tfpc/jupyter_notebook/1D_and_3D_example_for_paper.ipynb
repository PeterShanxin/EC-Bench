```python
import sys
sys.path.append("../")
import warnings
warnings.filterwarnings('ignore')
from utils.utils_jupyter_notebook import dict_of_amino_acid,show_3D_structure,predict_ec_and_scores_from_sequence,print_color_text,get_catalytic_res_from_sequence
import numpy as np
import nglview as nv
import pandas as pd
```

```python
#Decallage entre la séquence pdb et la séquence uniprot, Attention parfois c'est plus(+) qu'un simple décalage et 
#il faut refaire l'alignement, ici on a de la chance sur les exemples séléctionés.
dico_num_ex_decalage = {1 : 0,2:1,3:1,4:20}
inverse_dict_of_amino_acid = {value:key for key,value in dict_of_amino_acid.items()}
```

```python
def affiche_example(row):
    print("Numero example :",row["numero"])
    print("Description :",row["description"])
    print("PR_AUC",row["PR_AUC"])
    sequence = row["sequence"]
    print("The sequence size is :",len(sequence))
    decalage = dico_num_ex_decalage[row["numero"]]
    ec_number, res_scores,value_max = predict_ec_and_scores_from_sequence(sequence)
    top10 = np.flip(np.argsort(res_scores))[:10]
    print("Top 10 scores in uniprot indices:",[inverse_dict_of_amino_acid[sequence[ind]]+str(ind+1) for ind in top10])
    print("Top 10 scores map to pdb indices:",[inverse_dict_of_amino_acid[sequence[ind]]+str(ind+1-decalage) for ind in top10])
    indices_catalytic_res_pdb,pdb_id,chain_pdb,assembly_pdb,_ = get_catalytic_res_from_sequence(sequence, "pdb")
    indices_catalytic_res_uniprot,uniprot_id,_,_,binding_site_uniprot = get_catalytic_res_from_sequence(sequence, "uniprot")
    print("There are",len(indices_catalytic_res_pdb),"catalytic residues")
    print("Our model predict that your sequence are of the class : ",ec_number)
    print("indices_catalytic_res_pdb :",indices_catalytic_res_pdb)
    print("indices_catalytic_res_uniprot :",indices_catalytic_res_uniprot)
    print("Binding site id on uniprot map to pdb id :",list(np.array(binding_site_uniprot)-decalage))
    #assert indices_catalytic_res_pdb==indices_catalytic_res_uniprot
    cat_res = np.zeros(len(sequence))
    cat_res[np.array(indices_catalytic_res_uniprot)-1]=1
    print_color_text(sequence,cat_res,composante="green")
    print_color_text(sequence,res_scores,composante="red")
    print("Uniprot id:",uniprot_id)
    print("pdb_id :",pdb_id)
    print("chain_pdb :",chain_pdb)
    print("assembly_pdb :",assembly_pdb)
    for ind_cat in indices_catalytic_res_pdb:
        print("Rank interpretability scores :", np.where(np.flip(np.argsort(res_scores))==ind_cat))
   
    view = show_3D_structure(sequence,res_scores,pdb_id,chain_pdb,assembly_pdb,indices_catalytic_res_pdb,show_catalytic_res=True,decalage_uniprot_pdb=decalage)
    return view
```
